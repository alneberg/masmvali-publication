```python
ALL_STRATEGIES_ALPHA = ['velvetnoscaf','velvetscaf','velvetnoscafminimus2','velvetnoscafnewbler','velvetnoscafbambus2','velvetnoscafminimus2bambus2','velvetnoscafnewblerbambus2',
         'metavelvetnoscaf','metavelvetscaf','metavelvetnoscafminimus2','metavelvetnoscafnewbler','metavelvetnoscafbambus2','metavelvetnoscafminimus2bambus2','metavelvetnoscafnewblerbambus2',
         'raynoscaf','rayscaf','raynoscafminimus2','raynoscafnewbler', 'raynoscafbambus2', 'raynoscafminimus2bambus2', 'raynoscafnewblerbambus2']
        
POSTFIXES = [ '/noscaf/noscaf_X/ma-contigs.fa', 
              '/scaf/scaf_X/ma-scaffolds.fa',
              '/noscaf/minimus2/ma-merge.fa',
              '/noscaf/newbler/ma-merge.fa',
              '/noscaf/noscaf_X/bambus2/bambus2.scaffold.linear.fasta',
              '/noscaf/minimus2/bambus2/bambus2.scaffold.linear.fasta',
              '/noscaf/newbler/bambus2/bambus2.scaffold.linear.fasta' ]
                
FILEPATHD =  dict(zip(ALL_STRATEGIES_ALPHA,
                 [ item for sublist in
                 [ [ '/velvet' + pf for pf in POSTFIXES ],
                   [ '/metavelvet' + pf for pf in POSTFIXES ],
                   [ '/ray' + pf for pf in POSTFIXES ] ]
                  for item in sublist ]))

def alphafy_strat(strategy):
    alpha_strat = re.sub("scaf[0-9]{2}", "scaf", strategy)
    if alpha_strat not in FILEPATHD:
       raise(Exception("Invalid strategy: " + alpha_strat))
    else:
        return alpha_strat
        
def has_kmer(strategy):
    # Should consist of 2 numbers and should not be 22 (this is minimus2bambus2)
    return len(re.sub("[a-z]", "", strategy)) > 1 and int(re.sub("[a-z]", "", strategy)) != 22
    

def get_kmer(strategy):
    # Only take first two numbers, the 2 in bambus2 can add a number
    numbers = re.sub("[a-z]", "", strategy)
    return int(numbers[0:2])

def get_metassemble_assembly_path(d, strategy):
    d = d.rstrip('/')
    
    # Take only name without kmer number in it
    fp = FILEPATHD[alphafy_strat(strategy)]
        
    # Change X to requested kmer size
    if re.search("X", fp):
        fp = re.sub("X", str(get_kmer(strategy)), fp)
        
    fp = d + '/assemblies' + fp
    if not os.path.isfile(fp):
        warnings.warn("Strategy " + strategy + "'s fasta file does not exist: " + fp)
    
    return fp
    
def get_metassemble_assembly_validation(d, strategy, refset):
    fp = get_metassemble_assembly_path(d, strategy)
    return validation.AssemblyValidation(fp, "/".join(fp.split("/")[:-1]) + '/val/nucmer.coords', referenceset=refset)

# returns numpy array of asm-stats.tsv
def get_metassemble_assemstats(d, strategy, MISSING_VALUE="N/A", DELIMITER="\t"):
    fp = get_metassemble_assembly_path(d, strategy)
    a = np.genfromtxt("/".join(fp.split("/")[:-1]) + '/val/asm-stats.tsv', names=True, dtype=None, missing_values=MISSING_VALUE, usemask=True, delimiter=DELIMITER)
    return a
```

```python
MIN_KMER=21
MAX_KMER=75
STEPSIZE=2

def add_kmers(name, postfix="", rlist=range(MIN_KMER, MAX_KMER, STEPSIZE)):
    return [name + str(n) for n in rlist ]
    

asm_sets = {}
asm_sets["contig"] = add_kmers("velvetnoscaf") + add_kmers("raynoscaf") + add_kmers("metavelvetnoscaf")
asm_sets["scaf"] = add_kmers("velvetscaf") + add_kmers("rayscaf") + add_kmers("metavelvetscaf")
asm_sets['contig-merge'] = [n for n in ALL_STRATEGIES_ALPHA if n.endswith('newbler') or n.endswith('minimus2')]
asm_sets['contig-scaf'] = [a + "bambus2" for a in asm_sets["contig"]]
asm_sets['contig-merge-scaf'] = [a + "bambus2" for a in asm_sets["contig-merge"]]

print asm_sets['contig-merge']
print [alphafy_strat(asm) for asm in asm_sets['contig-merge']]
```

```python
PLOTD = dict(zip(ALL_STRATEGIES_ALPHA,
              ['k.','kv','ko','kd','k^','k*','kD',
               'g.','gv','go','gd','g^','g*','gD',
               'r.','rv','ro','rd','r^','r*','rD']
              )
          )
        
# Change figure size
rcParams['figure.figsize'] = 8, 6
rcParams['figure.dpi'] = 100
rcParams['savefig.dpi'] = 100

def plot_purity(directory, strategies, refset, legend_loc='lower right'):
    ax = subplot(1,1,1)
    for strategy in strategies:
        try:
            asmstats = get_metassemble_assemstats(directory, strategy)
        except IOError:
            warnings.warn("asmstats of " + strategy + " does not exist")
            continue
        if has_kmer(strategy):
            markersize = int(get_kmer(strategy) / float(MAX_KMER) * 15)
        else:
            markersize = 15
        astrat = alphafy_strat(strategy)
        plot(asmstats["l50"], asmstats["global_purity"], PLOTD[astrat], markersize=markersize, label=astrat)
    
    # Labels
    plt.xlabel('L50')
    plt.ylabel('Global Purity')
    
    # Legend
    handles, labels = ax.get_legend_handles_labels()
    # One label for each assembly strategy and one point shown in the legend
    labelsu = list(set(labels))
    handlesu = [handles[len(labels) - 1 - labels[::-1].index(l)] for l in labelsu]
    ax.legend(handlesu, labelsu, numpoints=1, loc=legend_loc)
    
    ax.set_ylim([0.5, 1.0])
    ax.set_xlim([0, 33000])
```
