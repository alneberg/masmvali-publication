## Compare contig purity

```python
%matplotlib inline
mpl.rcParams['figure.figsize'] = 16, 12

cmp_purities = ["velvetnoscaf31", "raynoscaf31", "metavelvetnoscaf31", 
                "velvetnoscafnewbler", "raynoscafnewbler", "metavelvetnoscafnewbler",
                "velvetnoscafminimus2", "raynoscafminimus2", "metavelvetnoscafminimus2"]
f, axes = plt.subplots(3, 3, sharex=True, sharey=True)

for i, s in enumerate(cmp_purities):
    ax = axes[i/3][i%3]
    ax.set_title(s)
    avp = m_even.avp.__dict__[s]
    ax.plot(avp.contig_purity["contig_length"], avp.contig_purity["max_aln_purity"], '.', linewidth=3.0)


# Set common labels
f.text(0.5, 0.08, 'Contig length', ha='center', va='center')
f.text(0.08, 0.5, 'Purity', ha='center', va='center', rotation='vertical')
```

## Metagenome coverage

```python
mpl.rcParams['figure.figsize'] = 10, 8

mg_length = np.sum(m_even.avp.velvetnoscaf31.genome_contig_coverage["genome_length"])
f, ax1 = plt.subplots(1, 1, sharex=True, sharey=True)
#ax1.set_xlim([0, 600000])
#ax1.set_title("filtered-refs-filtered-pairs-noscaf")
ax1.plot(m_even.avp.velvetnoscaf31.contig_metagenome_coverage_purest_single_aln["cut_off"], 
         m_even.avp.velvetnoscaf31.contig_metagenome_coverage_purest_single_aln["metagenome_cov_percentage"],
         'k', linewidth=2.0, label='velvetnoscaf31')
ax1.plot(m_even.avp.raynoscaf31.contig_metagenome_coverage_purest_single_aln["cut_off"], 
         m_even.avp.raynoscaf31.contig_metagenome_coverage_purest_single_aln["metagenome_cov_percentage"],
         'r', linewidth=2.0, label='raynoscaf31')
ax1.plot(m_even.avp.metavelvetnoscaf31.contig_metagenome_coverage_purest_all_aln["cut_off"], 
         m_even.avp.metavelvetnoscaf31.contig_metagenome_coverage_purest_all_aln["metagenome_cov_percentage"],
         'g', linewidth=2.0, label='metavelvetnosaf31')

for ax in [ax1]:
    ax.set_xscale('log')
plt.gca().invert_xaxis()

for ax in [ax1]:
    ax.set_ylim([0,100])
    ax.set_xlim([10**6,0])

plt.ylabel('Metagenome coverage (%)')
plt.xlabel('Contig/Scaffold length')
# Legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, numpoints=1, loc='upper left')
```
