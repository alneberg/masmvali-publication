```python
ALL_STRATEGIES_ALPHA = ['velvetnoscaf','velvetscaf','velvetnoscafminimus2','velvetnoscafnewbler','velvetnoscafbambus2','velvetnoscafminimus2bambus2','velvetnoscafnewblerbambus2',
         'metavelvetnoscaf','metavelvetscaf','metavelvetnoscafminimus2','metavelvetnoscafnewbler','metavelvetnoscafbambus2','metavelvetnoscafminimus2bambus2','metavelvetnoscafnewblerbambus2',
         'raynoscaf','rayscaf','raynoscafminimus2','raynoscafnewbler', 'raynoscafbambus2', 'raynoscafminimus2bambus2', 'raynoscafnewblerbambus2']
        
POSTFIXES = [ '/noscaf/noscaf_X/ma-contigs.fa', 
              '/scaf/scaf_X/ma-scaffolds.fa',
              '/noscaf/minimus2/ma-merge.fa',
              '/noscaf/newbler/ma-merge.fa',
              '/noscaf/noscaf_X/bambus2/bambus2.scaffold.linear.fasta',
              '/noscaf/minimus2/bambus2/bambus2.scaffold.linear.fasta',
              '/noscaf/newbler/bambus2/bambus2.scaffold.linear.fasta' ]
                
FILEPATHD =  dict(zip(ALL_STRATEGIES_ALPHA,
                 [ item for sublist in
                 [ [ '/velvet' + pf for pf in POSTFIXES ],
                   [ '/metavelvet' + pf for pf in POSTFIXES ],
                   [ '/ray' + pf for pf in POSTFIXES ] ]
                  for item in sublist ]))

def alphafy_strat(strategy):
    alpha_strat = re.sub("scaf[0-9]{2}", "scaf", strategy)
    if alpha_strat not in FILEPATHD:
       raise(Exception("Invalid strategy: " + alpha_strat))
    else:
        return alpha_strat
        
def has_kmer(strategy):
    # Should consist of 2 numbers and should not be 22 (this is minimus2bambus2)
    return len(re.sub("[a-z]", "", strategy)) > 1 and int(re.sub("[a-z]", "", strategy)) != 22
    

def get_kmer(strategy):
    # Only take first two numbers, the 2 in bambus2 can add a number
    numbers = re.sub("[a-z]", "", strategy)
    return int(numbers[0:2])

def get_metassemble_assembly_path(d, strategy):
    d = d.rstrip('/')
    
    # Take only name without kmer number in it
    fp = FILEPATHD[alphafy_strat(strategy)]
        
    # Change X to requested kmer size
    if re.search("X", fp):
        fp = re.sub("X", str(get_kmer(strategy)), fp)
        
    fp = d + '/assemblies' + fp
    if not os.path.isfile(fp):
        warnings.warn("Strategy " + strategy + "'s fasta file does not exist: " + fp)
    
    return fp

def get_metassemble_assembly_validation_parser(d, strategy, missing_value="N/A"):
    fp = get_metassemble_assembly_path(d, strategy)
    return validation.AssemblyValidationParser("/".join(fp.split("/")[:-1]) + '/val', missing_value)
    
def get_metassemble_assembly_validation(d, strategy, refset):
    fp = get_metassemble_assembly_path(d, strategy)
    return validation.AssemblyValidation(fp, "/".join(fp.split("/")[:-1]) + '/val/nucmer.coords', referenceset=refset)

# returns numpy array of asm-stats.tsv
def get_metassemble_assemstats(d, strategy, MISSING_VALUE="N/A", DELIMITER="\t"):
    fp = get_metassemble_assembly_path(d, strategy)
    a = np.genfromtxt("/".join(fp.split("/")[:-1]) + '/val/asm-stats.tsv', names=True, dtype=None, missing_values=MISSING_VALUE, usemask=True, delimiter=DELIMITER)
    return a
```

```python
MIN_KMER=21
MAX_KMER=75
STEPSIZE=2

def add_kmers(name, postfix="", rlist=range(MIN_KMER, MAX_KMER, STEPSIZE)):
    return [name + str(n) for n in rlist ]

def get_max_l50(d, strategies):
    def get_l50(s):
        try:
            rv = get_metassemble_assemstats(d, s)["l50"]
        except IOError:
            warnings.warn("asmstats of " + s + " does not exist")
            rv = 0
        return rv
            
    return max(strategies, key=get_l50)

asm_sets = {}
for k in DIRS:
    asm_sets[k] = {}
    asm_sets[k]["contig"] = [get_max_l50(DIRS[k], add_kmers(s)) for s in ["raynoscaf", "velvetnoscaf", "metavelvetnoscaf"]]
    asm_sets[k]["scaf"] = [get_max_l50(DIRS[k], add_kmers(s)) for s in ["rayscaf", "velvetscaf", "metavelvetscaf"]]
    asm_sets[k]['contig-merge'] = [n for n in ALL_STRATEGIES_ALPHA if n.endswith('newbler') or n.endswith('minimus2')]
    asm_sets[k]['contig-scaf'] = [get_max_l50(DIRS[k], [a + "bambus2" for a in add_kmers(s)]) for s in ["raynoscaf", "velvetnoscaf", "metavelvetnoscaf"]]
    asm_sets[k]['contig-merge-scaf'] = [a + "bambus2" for a in asm_sets[k]["contig-merge"]]

for k in DIRS:
    print k
    print asm_sets[k]['contig']
    print asm_sets[k]['contig-scaf']
```

```python
PLOTD = dict(zip(ALL_STRATEGIES_ALPHA,
              ['k-','k-','k--','k-','k-','k--','k-',
               'g-','g-','g--','g-','g-','g--','g-',
               'r-','r-','r--','r-','r-','r--','r-']
              )
          )
        
# Change figure size
mpl.rcParams['figure.figsize'] = 8, 6
mpl.rcParams['figure.dpi'] = 100
mpl.rcParams['savefig.dpi'] = 100

def plot_length(directory, strategies, cut_off=100):
    ax = plt.subplot(1,1,1)
    for strategy in strategies:
        fp = get_metassemble_assembly_path(directory, strategy)
        if os.path.isfile(fp):
            avp = validation.AssemblyValidationParser("/".join(fp.split("/")[:-1]) + '/val')
        
        # Plot contig length vs sum of bases
        astrat = alphafy_strat(strategy)
        cl = avp.contig_lengths
        plt.plot(cl["length"][cl["length"] > cut_off], cl["sum"][cl["length"] > cut_off], PLOTD[astrat], label=astrat, linewidth=2.0)
        
        # Add L50
        sum_cut_off = sum(avp.contig_lengths["length"][avp.contig_lengths["length"] > cut_off])
        l50 = avp.contig_lengths["length"][avp.contig_lengths["length"] > cut_off][avp.contig_lengths["sum"] >= sum_cut_off/2][0]
        l50, n50, sum_50_cut_off, sum_cut_off = avp.get_nx_stats(50)
        plt.plot(l50, sum_50_cut_off, PLOTD[astrat][0] + "o", markersize=10)
    ax.set_xscale('log')
    ax.set_ylim([0,METAGENOME_LENGTH])
    plt.gca().invert_xaxis()
    # Labels
    plt.xlabel('Contig/Scaffold length')
    plt.ylabel('Sum of bases')
    # Legend
    handles, labels = ax.get_legend_handles_labels()
    #labelsu = list(set(labels))
    #handlesu = [handles[len(labels) - 1 - labels[::-1].index(l)] for l in labelsu]
    ax.legend(handles, labels, loc='upper left') # or lower right
    #ax.set_yscale('log')
    ax.set_ylim([0, METAGENOME_LENGTH])
    ax.set_xlim([10**6,10**2])
```

```python
def plot_length_vs_metagenome_cov(directory, strategies, refset, cut_off=100):
    
    #lengthsa = np.array([item for sublist in map(lambda x: range(10**x,10**(x+1),10**x),range(2,7)) for item in sublist]) 
    ax = plt.subplot(1,1,1)
    for strategy in strategies:
        fp = get_metassemble_assembly_path(directory, strategy)
        if os.path.isfile(fp):
            avp = validation.AssemblyValidationParser("/".join(fp.split("/")[:-1]) + '/val')
        
            cmgc = avp.contig_metagenome_coverage
            astrat = alphafy_strat(strategy)
            plot(cmgc["cut_off"], cmgc["metagenome_cov_percentage"], PLOTD[astrat], label=astrat, linewidth=2.0)
    ax.set_xscale('log')
    plt.gca().invert_xaxis()
    # Labels
    plt.ylabel('Metagenome coverage (%)')
    plt.xlabel('Contig/Scaffold length')
    # Legend
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, labels, numpoints=1, loc='upper left')
    
    ax.set_ylim([0,100])
    ax.set_xlim([10**6,0])
```
