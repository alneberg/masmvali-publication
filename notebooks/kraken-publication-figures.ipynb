### Find LCA rank between all taxonomy ids

```python
taxonomy = kraken.Taxonomy(nodes_file, tax_names_file, tids_file)
```

```python
krakens = [kraken.Kraken(k, taxonomy) for k in kraken_files]
```

```python
mpl.rcParams['figure.figsize'] = 6, 9
asms = ["velvetnoscaf31", "raynoscaf31", "raynoscafnewbler"]
panels = ["A","B","C"]
f, axes = plt.subplots(3, 1)
for i, s in enumerate(asms):
    ax = axes[i]
    #ax.text(0.05, 1e08, panels[i], fontweight='bold', size='xx-large', ha='center', va='center')
    #ax.text(0.09, 1e08, s, size='large', ha='left', va='center')
    ax.text(1, 85, panels[i], fontweight='bold', size='xx-large', ha='center', va='center')
    ax.text(2, 85, s, size='large', ha='left', va='center')
    (krakens[i].lca_rank_kmer_count_sort.sum() / krakens[i].lca_rank_kmer_count_sort.sum().sum() * 100.0).plot(ax=ax, kind='bar', ylim=[0,100])
    
for ax in axes.flatten()[:-1]:
    ax.set_xticklabels([])
    
sns.despine()
plt.savefig("/media/milou/glob/github/masmvali-publication/plos/figures/Figure_3.eps", 
            bbox_extra_artists=(ax.yaxis,), bbox_inches='tight')
```

```python
mpl.rcParams['figure.figsize'] = 6, 9
asms = ["velvetnoscaf31", "raynoscaf31", "raynoscafnewbler"]
panels = ["A","B","C"]
f, axes = plt.subplots(3, 1)
for i, s in enumerate(asms):
    ax = axes[i]
    cpurity = pd.read_csv(contig_purity_files[i], sep="\t").sort("contig").reset_index().set_index("contig")
    cpurity_lca = cpurity.ix[:,["max_aln_strain","max_aln_purity","contig_length"]].join(krakens[i].lca_rank_kmer_count_sort, how='right')
    pure = cpurity_lca.max_aln_purity == 1
    impure = cpurity_lca.max_aln_purity < 1
    chimeric_kmers = cpurity_lca.ix[:,range(3,12)].sum(axis=1) > 0
    error_kmers = cpurity_lca.unknown > 0
    v3 = venn3(ax=ax, subsets=(
           (impure & ~chimeric_kmers & ~error_kmers).sum() * 10000 / impure.sum() / 100.0,
           (impure &  chimeric_kmers & ~error_kmers).sum() * 10000 / impure.sum() / 100.0,
           0,
           (impure & ~chimeric_kmers &  error_kmers).sum() * 10000 / impure.sum() / 100.0,
           0,
           (impure &  chimeric_kmers &  error_kmers).sum() * 10000 / impure.sum() / 100.0,
           0),
           set_labels = ("Correct", "Erroneous", "Chimeric"), alpha=0.4, set_colors=('g','r','b'))
    #ax.text(0.1, 0.5, panels[i], fontweight='bold', size='xx-large', ha='center', va='center')
    ax.set_title(panels[i], fontweight='bold', size='xx-large')
    plt.savefig("/media/milou/glob/github/masmvali-publication/plos/figures/Figure_4.pdf")
```
