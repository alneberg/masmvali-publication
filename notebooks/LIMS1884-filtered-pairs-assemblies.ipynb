## Compare Contig Purity
### old references

````python
mpl.rcParams['figure.figsize'] = 16, 12

cmp_purities = ["velvetnoscaf31", "raynoscaf31", "metavelvetnoscaf31", 
                "velvetnoscafnewbler", "raynoscafnewbler", "metavelvetnoscafnewbler",
                "velvetnoscafminimus2", "raynoscafminimus2", "metavelvetnoscafminimus2"]
f, axes = subplots(3, 3, sharex=True, sharey=True)

for i, s in enumerate(cmp_purities):
    ax = axes[i/3][i%3]
    ax.set_title(s)
    avp = m_even_old.avp.__dict__[s]
    ax.plot(avp.contig_purity["contig_length"], avp.contig_purity["max_aln_purity"], '.', linewidth=3.0)


# Set common labels
f.text(0.5, 0.08, 'Contig length', ha='center', va='center')
f.text(0.08, 0.5, 'Purity', ha='center', va='center', rotation='vertical')
````

From these from these plots it would seem that velvet keeps the contigs purest, but it could also just be that velvet has the shortest contigs. There's definitely a lot of unpure contigs while using the original references and all pairs.
