## Metagenome coverage using different alignment filters

Since we are still not sure what would be the best way to calculate metagenome coverage, I have tried three different alignment filters:

- **purest-single** use one purest alignment per contig
- **purest-all** use all purest alignemtns per contig (in case there are multiple)
- **pure-only-single** use only 100% pure alignments, use only one in case there are multiple for a contig
- **pure-only-all**  use only 100% pure alignments, use all in case there are multiple for a contig
- **all** use all alignments regardless of purity

````python
mpl.rcParams['figure.figsize'] = 10, 8

mg_length = np.sum(m_even.avp.velvetnoscaf31.genome_contig_coverage["genome_length"])
f, ax1 = plt.subplots(1, 1, sharex=True, sharey=True)
#ax1.set_xlim([0, 600000])
ax1.set_title("filtered-refs-filtered-pairs-velvetnoscaf31")
ax1.plot(m_even.avp.velvetnoscaf31.contig_metagenome_coverage_purest_single_aln["cut_off"], 
         m_even.avp.velvetnoscaf31.contig_metagenome_coverage_purest_single_aln["metagenome_cov_percentage"],
         'g', linewidth=2.0, label='purest-single')
ax1.plot(m_even.avp.velvetnoscaf31.contig_metagenome_coverage_pure_only_single_aln["cut_off"], 
         m_even.avp.velvetnoscaf31.contig_metagenome_coverage_pure_only_single_aln["metagenome_cov_percentage"],
         'g--', linewidth=2.0, label='pure-only-single')
ax1.plot(m_even.avp.velvetnoscaf31.contig_metagenome_coverage_purest_all_aln["cut_off"], 
         m_even.avp.velvetnoscaf31.contig_metagenome_coverage_purest_all_aln["metagenome_cov_percentage"],
         'b', linewidth=2.0, label='purest-all')
ax1.plot(m_even.avp.velvetnoscaf31.contig_metagenome_coverage_pure_only_all_aln["cut_off"], 
         m_even.avp.velvetnoscaf31.contig_metagenome_coverage_pure_only_all_aln["metagenome_cov_percentage"],
         'b--', linewidth=2.0, label='pure-only-all')
ax1.plot(m_even.avp.velvetnoscaf31.contig_metagenome_coverage_all_aln["cut_off"], 
         m_even.avp.velvetnoscaf31.contig_metagenome_coverage_all_aln["metagenome_cov_percentage"],
         'k', linewidth=2.0, label='all')
ax1.plot(m_even.avp.velvetnoscaf31.contig_metagenome_coverage_all_aln["cut_off"], 
         m_even.avp.velvetnoscaf31.contig_metagenome_coverage_all_aln["assembly_bases"] * 100 / mg_length,
         'r', linewidth=2.0, label='bases in assembly')
ax1.plot(m_even_old.avp.velvetnoscaf31.contig_metagenome_coverage["cut_off"], 
         m_even_old.avp.velvetnoscaf31.contig_metagenome_coverage["metagenome_cov_percentage"],
         'y', linewidth=2.0, label='purest-single unfiltered pairs')

for ax in [ax1]:
    ax.set_xscale('log')
plt.gca().invert_xaxis()

for ax in [ax1]:
    ax.set_ylim([0,100])
    ax.set_xlim([10**6,0])

plt.ylabel('Metagenome coverage (%)')
plt.xlabel('Contig/Scaffold length')
# Legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, numpoints=1, loc='upper left')
````

For this assembly there is little difference between using all purest alignments (**purest-all**) or using only one of the purest alignments (**purest-single**). Using all alignments does result in quite a big difference (**all**), but one can see that it is mostly due to the same bases being aligned at multiple places instead of bases being "trapped" in their purest alignment since the line **bases in assembly** is overlapping with **purest-all**. The metagenome coverage is quite a bit better compared to the old results with unfiltered pairs and refs. It is unclear atm whether this is due to the filtering of the pairs or removement of the references.

## Investigate low purity and N/A contigs
### N/A contigs
There are still some contigs left that do not map anywhere. Most of them are really short contigs (< 100), they are currently not filtered by masmvali. There were only 4 contigs for velvetnoscaf31 that were not mapping. After aligning with blast:

```
$ cd /gulo/glob/inod/projects/masmvali-partdeux/reassembly-filtered-reads/Sample_1ng_even/metassemble/assemblies/velvet/noscaf/noscaf_31/val/blastNAcontigs/blast
$ blastn -db nt -num_threads 1 -max_target_seqs 1 -outfmt '6 std stitle' -query ../contigs-min500-purity-below-09.fa > contigs-min500-purity-below-09-megablast.tsv
```
Turned out that all of them aligned to Sulfurihydrogenibium sp. YO3AOP1. The nucmer minimum cluster parameter '-c 65' was too large for MUMmer to find the alignment. For all of them there would be only one match with exactly matching length -l 20 which was not big enough to be a cluster and would therefore not be extended (I assume). We could either only use larger contigs to mitigate this problem or change the minimum cluster parameter to something smaller for more sensitivity.

### Test different distributions for coverage mean over the genomes

````python
import scipy.stats
````

````python
%matplotlib inline
mpl.rcParams['figure.figsize'] = 10, 8

f, axes = plt.subplots(2, 2)
((osm, osr), (slope, intercept, r))  = scipy.stats.probplot(even_filt_ref["cov"], dist="norm", plot=axes[0][0])
axes[0][0].set_title("Norm")
axes[0][0].set_xlabel("")
((osm, osr), (slope, intercept, r))  = scipy.stats.probplot(even_filt_ref["cov"], dist="poisson", sparams=(even_filt_ref["cov"].mean(),), plot=axes[0][1])
axes[0][1].set_title("Poisson (param=mean)")
axes[0][1].set_xlabel("")
axes[0][1].set_ylabel("")
((osm, osr), (slope, intercept, r))  = scipy.stats.probplot(even_filt_ref["cov"], dist="t", sparams=(even_filt_ref["cov"].mean(),), plot=axes[1][0])
axes[1][0].set_title("Student T (param=mean)")
((osm, osr), (slope, intercept, r))  = scipy.stats.probplot(even_filt_ref["cov"], dist="lognorm", sparams=(1,), plot=axes[1][1])
axes[1][1].set_title("Log norm(param=1)")
axes[1][1].set_xlabel("")
````
