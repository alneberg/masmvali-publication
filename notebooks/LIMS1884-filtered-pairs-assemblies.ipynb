## Metagenome coverage using different alignment filters

Since we are still not sure what would be the best way to calculate metagenome coverage, I have tried three different alignment filters:

- **purest-single** use one purest alignment per contig
- **purest-all** use all purest alignemtns per contig (in case there are multiple)
- **all** use all alignments regardless of purity

````python
mpl.rcParams['figure.figsize'] = 10, 8

mg_length = np.sum(m_even.avp.velvetnoscaf31.genome_contig_coverage["genome_length"])
f, ax1 = subplots(1, 1, sharex=True, sharey=True)
#ax1.set_xlim([0, 600000])
ax1.set_title("filtered-refs-filtered-pairs-velvetnoscaf31")
ax1.plot(m_even.avp.velvetnoscaf31.contig_metagenome_coverage_purest_single_aln["cut_off"], 
         m_even.avp.velvetnoscaf31.contig_metagenome_coverage_purest_single_aln["metagenome_cov_percentage"],
         'g', linewidth=2.0, label='purest-single')
ax1.plot(m_even.avp.velvetnoscaf31.contig_metagenome_coverage_purest_all_aln["cut_off"], 
         m_even.avp.velvetnoscaf31.contig_metagenome_coverage_purest_all_aln["metagenome_cov_percentage"],
         'b', linewidth=2.0, label='purest-all')
ax1.plot(m_even.avp.velvetnoscaf31.contig_metagenome_coverage_all_aln["cut_off"], 
         m_even.avp.velvetnoscaf31.contig_metagenome_coverage_all_aln["metagenome_cov_percentage"],
         'k', linewidth=2.0, label='all')
ax1.plot(m_even.avp.velvetnoscaf31.contig_metagenome_coverage_all_aln["cut_off"], 
         m_even.avp.velvetnoscaf31.contig_metagenome_coverage_all_aln["assembly_bases"] * 100 / mg_length,
         'r', linewidth=2.0, label='bases in assembly')

for ax in [ax1]:
    ax.set_xscale('log')
plt.gca().invert_xaxis()

for ax in [ax1]:
    ax.set_ylim([0,100])
    ax.set_xlim([10**6,0])

plt.ylabel('Metagenome coverage (%)')
plt.xlabel('Contig/Scaffold length')
# Legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, numpoints=1, loc='upper left')
````

For this assembly there is little difference between using all purest alignments (**purest-all**) or using only one of the purest alignments (**purest-single**). Using all alignments does result in quite a big difference (**all**), but one can see that it is mostly due to the same bases being aligned at multiple places instead of bases being "trapped" in their purest alignment since the line (**bases in assembly**) is overlapping with **purest-all**.
