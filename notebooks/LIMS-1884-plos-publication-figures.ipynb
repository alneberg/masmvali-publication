## Main

```python
mpl.rcParams['figure.figsize'] = 6, 9
mpl.rc('text', usetex=False)
mpl.rcParams['legend.fancybox'] = False

cmp_purities = ["velvetnoscaf31", "raynoscaf31", "metavelvetnoscaf31"]
panels = ["1A","1B","1C"]
labels = ["$>10^2$","$>10^3$", "$>10^4$", "$>10^5$"]
f, axes = plt.subplots(3, 1, sharex=True, sharey=True)
for i, s in enumerate(cmp_purities):
    ax = axes[i]
    avp = m_even.avp.__dict__[s]
    
    # interval lengths
    len_intervals = []
    len_weights = []
    ivs = [(int(1e2), int(1e3)), (int(1e3), int(1e4)), (int(1e4), int(1e5)), (int(1e5),float("inf"))]
    for l, h in ivs:
        leni = (np.array(avp.contig_purity["contig_length"] >= l) & np.array(avp.contig_purity["contig_length"] < h))
        len_intervals.append(list(avp.contig_purity[leni]["max_aln_purity"]))
        len_weights.append(list(avp.contig_purity[leni]["contig_length"]))
    
    ax.hist(len_intervals, 
         weights=len_weights,
         range=[0.0,1.0], log=1,
         label=labels,
         color=sns.dark_palette("palegreen", len(ivs)))
    ax.set_ylim(1,1e09)
    #ax.set_title("1A. " + s)
    ax.text(0.05, 1e08, panels[i], fontweight='bold', size='xx-large', ha='center', va='center')
    ax.text(0.09, 1e08, s, size='large', ha='left', va='center')
    #ax.yaxis.grid(True, c='white')
    #ax.set_axisbelow(False)
    #sns.despine()
    
# Set common labels
#f.text(0.5, 0.95, "Even Community", va='center', ha='center', size='xx-large')
t1 = f.text(0.5, 0.07, 'Purity', ha='center', va='center', size='xx-large')
t2 = f.text(0.03, 0.5, 'Bases in contigs with purity $x$ (log)', ha='center', va='center', rotation='vertical', size='xx-large')

#axes[2].legend(title="Contig length", loc='lower left', ncol=1)
lgd = ax.legend(title="Contig length", bbox_to_anchor=(1.05, 0), loc='lower left', borderaxespad=0.)
#t2 = axes[0].get_legend().get_title()
#t2.set_family(t1.get_family())

plt.savefig("/media/milou/glob/github/masmvali-publication/plos/figures/Figure_1.eps", 
            bbox_extra_artists=(lgd, t1, t2), bbox_inches='tight')
```
