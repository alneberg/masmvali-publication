## QUAST for metagenomics
Quality Assessment Tool for Genome Assemblies (QUAST) is a tool that can help you assess the quality of an assembly with and without a reference genome. It has been updated to work with metagenomics since version 2.2. Apart from the excellent plots it outputs it also outputs the stats in several tsv files, which we will be looking at to make our own plots.

```python
import pandas as pd
import glob
import matplotlib as mpl
```

```python
# There's a separate folder for each reference genome with a report tsv file
reports = glob.glob("/media/milou//glob/projects/masmvali-partdeux/reassembly-filtered-reads/Sample_1ng_even/metassemble/assemblies/velvet/noscaf/noscaf_31/quast/quast_results/latest/ref_*_quast_output/transposed_report.tsv")
reports.sort()
```

```python
reportsdf = pd.read_csv(reports[0], sep='\t')
for r in reports[1:]:
    reportsdf = reportsdf.append(pd.read_csv(r, sep='\t'))
reportsdf['ref'] = [r[r.find('ref_') + len('ref_'):r.find('_quast_output/')] for r in reports]
reportsdf.set_index('ref', inplace=True)
```

```python
%matplotlib inline
mpl.rcParams['figure.figsize'] = 10, 6
reportsdf.ix[:,['# local misassemblies', '# misassembled contigs', '# misassemblies']].plot(kind='bar', yticks=range(5))
```

```python
has_misasm = (reportsdf['# local misassemblies'] > 0) | (reportsdf['# misassembled contigs'] > 0) | (reportsdf['# misassemblies'] > 0)
%matplotlib inline
mpl.rcParams['figure.figsize'] = 10, 6
reportsdf[has_misasm].ix[:,['# local misassemblies', '# misassembled contigs', '# misassemblies']].plot(kind='bar', yticks=range(5))
```
