```python
N_paths = 5
N_steps = 100

x = np.linspace(0, 10, 100)
y = 0.1 * (np.random.random((N_paths, N_steps)) - 0.5)
y = y.cumsum(1)

fig, ax = plt.subplots()
labels = ["a", "b", "c", "d", "e"]
line_collections = ax.plot(x, y.T, lw=4, alpha=0.2)
interactive_legend = plugins.InteractiveLegendPlugin(line_collections, labels)
plugins.connect(fig, interactive_legend)

mpld3.display()
```

```python
pths = sorted(np.unique(m_even.avp.velvetnoscafminimus2.contig_metagenome_coverage_purest_single_aln_purity_thresholds["min_purity"]))
sns.set_style("darkgrid")
#sns.set_style("white")
colors = sns.dark_palette("skyblue", len(pths))
#sns.set()

f, ax = plt.subplots(1, 1,)
#m_even.avp.velvetnoscafminimus2.contig_metagenome_coverage_purest_single_aln_purity_thresholds["min_purity"]
line_collections = []
for i, p in enumerate(pths):
    has_purity = np.array(m_even.avp.velvetnoscafminimus2.contig_metagenome_coverage_purest_single_aln_purity_thresholds["min_purity"] == p)
    line_collections.append(ax.plot(m_even.avp.velvetnoscafminimus2.contig_metagenome_coverage_purest_single_aln_purity_thresholds[has_purity]["cut_off"],
            m_even.avp.velvetnoscafminimus2.contig_metagenome_coverage_purest_single_aln_purity_thresholds[has_purity]["metagenome_cov_percentage"],
            linewidth=4, color=colors[i], label=p, alpha=(0.2)))


ax.set_xscale('log')
ax.invert_xaxis()
ax.set_ylim([0,100])
ax.set_xlim([10**6,0])
ax.set_title("minimusvelvetnoscaf31")
ax.set_ylabel("Metagenome coverage (%)", size='xx-large')
ax.set_xlabel("Contig/Scaffold length", size='xx-large')
f.set_size_inches(10.5,6)
plugins.connect(f, plugins.InteractiveLegendPlugin(line_collections, pths))
mpld3.save_html(f, "purity-thresholds.html")
mpld3.display()
```
