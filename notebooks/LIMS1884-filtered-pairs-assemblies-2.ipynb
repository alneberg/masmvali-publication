### Using coverage and length to extract pure contigs only

```python
rns31cov = pd.read_csv("/media/milou/glob/projects/masmvali-partdeux/reassembly-filtered-reads/Sample_1ng_even/metassemble/assemblies/ray/noscaf/noscaf_31/bambus2/contigs_pair-smds.coverage.percontig", 
            sep="\t").sort("contig").set_index("contig")
rns31cov.head()
```

```python
rns31pur = np.sort(m_even.avp.raynoscaf31.contig_purity, order='contig')
```

```python
f, ax = plt.subplots(1, 1)
goodcov = np.array((rns31cov["cov_mean_sample_0"] > 15) & (rns31cov["percentage_covered_sample_0"] > 10))
longc = rns31pur["contig_length"] >= 1000
ax.hist([rns31pur[longc & ~goodcov]["max_aln_purity"],
         rns31pur[longc & goodcov]["max_aln_purity"]],
         weights=[rns31pur[longc & ~goodcov]["contig_length"],
                  rns31pur[longc & goodcov]["max_aln_purity"]], log=1, range=[0.0,1.0], label=["bad coverage","good coverage"],
         align='mid')
ax.legend()
ax.set_title("raynoscaf31")
ax.set_ylabel("log # bases in contig with purity (x)")
ax.set_xlabel("purity")
ax.set_ylim(1,1e09)
ax.set_xlim(0.0, 1.0)
```

Using only coverage and length it is rather hard to select pure contigs. You lose a lot of contigs that have excellent purity. Maybe changing mapping parameters could improve this. Or using FRCurve to determine features per contig instead of just looking at mean coverage.

## Metagenome coverage

```python
mpl.rcParams['figure.figsize'] = 10, 8

mg_length = np.sum(m_even.avp.velvetnoscaf31.genome_contig_coverage["genome_length"])
f, axes = plt.subplots(2, 2, sharex=True, sharey=True)
#ax1.set_xlim([0, 600000])
#ax1.set_title("filtered-refs-filtered-pairs-noscaf")
def plot_metagenome_cov(ax, masm, strategies, category, colors):
    for i, s in enumerate(strategies):
        ax.plot(masm.avp.__dict__[s].contig_metagenome_coverage_purest_single_aln["cut_off"],
                masm.avp.__dict__[s].contig_metagenome_coverage_purest_single_aln["metagenome_cov_percentage"],
                colors[i], linewidth=2.0, label=s)
    ax.set_title(category)

plot_metagenome_cov(axes[0][0], m_even, ['velvetnoscaf31', 'raynoscaf31', 'metavelvetnoscaf31'], 
                    'noscaf', ['k','r','g'])
plot_metagenome_cov(axes[0][1], m_even, ['velvetscaf31', 'rayscaf31', 'metavelvetscaf31'], 
                    'scaf', ['k','r','g'])
plot_metagenome_cov(axes[1][0], m_even, ['velvetnoscafminimus2', 'raynoscafminimus2', 'metavelvetnoscafminimus2'], 
                    'minimus2', ['k','r','g'])
plot_metagenome_cov(axes[1][1], m_even, ['velvetnoscafnewbler', 'raynoscafnewbler', 'metavelvetnoscafnewbler'], 
                    'newbler', ['k','r','g'])



for ax in axes.flatten():
    ax.set_ylim([0,100])
    ax.set_xlim([10**6,0])
    
for ax in axes.flatten():
    ax.set_xscale('log')
    ax.invert_xaxis()
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, labels, numpoints=1, loc='upper left')

#plt.ylabel('Metagenome coverage (%)')
#plt.xlabel('Contig/Scaffold length')

# Legend
# Set common labels
f.text(0.5, 0.08, 'Contig/Scaffold length', ha='center', va='center')
f.text(0.08, 0.5, 'Metagenome coverage (%)', ha='center', va='center', rotation='vertical')
```

As we can see, metavelvet and ray provide way more metagenome coverage but at a cost of purity. It would actually be interesting to also do these plots with only pure contigs, so one can see the difference. The number of bases in impure contigs compared to the pure ones are actually quite low, so I'm guessing there is little difference to be seen.
