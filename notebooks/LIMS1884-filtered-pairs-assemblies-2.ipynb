### Extracting pure contigs only
#### Using coverage and length

````python
rns31cov = pd.read_csv("/media/milou/glob/projects/masmvali-partdeux/reassembly-filtered-reads/Sample_1ng_even/metassemble/assemblies/ray/noscaf/noscaf_31/bambus2/contigs_pair-smds.coverage.percontig", 
            sep="\t").sort("contig").set_index("contig")
rns31cov.head()
````

````python
rns31pur = np.sort(m_even.avp.raynoscaf31.contig_purity, order='contig')
````

````python
f, ax = plt.subplots(1, 1)
goodcov = np.array((rns31cov["cov_mean_sample_0"] > 60) & (rns31cov["percentage_covered_sample_0"] > 90))
longc = rns31pur["contig_length"] >= 1000
ax.hist([rns31pur[longc & ~goodcov]["max_aln_purity"],
         rns31pur[longc & goodcov]["max_aln_purity"]],
         weights=[rns31pur[longc & ~goodcov]["contig_length"],
                  rns31pur[longc & goodcov]["contig_length"]], log=1, range=[0.0,1.0], label=["bad coverage","good coverage"],
         align='mid')
ax.legend()
ax.set_title("raynoscaf31")
ax.set_ylabel("log # bases in contig with purity (x)")
ax.set_xlabel("purity")
ax.set_ylim(1,1e09)
ax.set_xlim(0.0, 1.0)
````

Using only coverage and length it is tricky select pure contigs. You lose a lot of contigs that have excellent purity. Maybe changing mapping parameters could improve this. Or using FRCurve to determine features per contig instead of just looking at mean coverage.

#### Extracting pure contigs with FRC
[FRCurve](https://github.com/vezzi/FRC_align) is a referenceless assembly validation tool. It gives you a lot of different ouput files as can be seen in the readme on the github page. We are interested in getting the number of features indicating a wrong assembly per contig. I ran the program for raynoscaf31 with:
```
~/glob/src/FRC_align/bin/FRC --pe-sam ../bambus2/contigs_pair-smds.bam --genome-size 169669119
```

````python
cfeat = pd.read_csv("/media/milou/glob/projects/masmvali-partdeux/reassembly-filtered-reads/Sample_1ng_even/metassemble/assemblies/ray/noscaf/noscaf_31/frc/Features.txt",
                    sep=" ", names=['contig','feature','start','end']).sort("contig")
````

````python
cfeat_counts = cfeat.groupby('contig').count().ix[:,["feature"]]
````

````python
rns31purdf = pd.read_csv("/media/milou/glob/projects/masmvali-partdeux/reassembly-filtered-reads/Sample_1ng_even/metassemble/assemblies/ray/noscaf/noscaf_31/val/contig-purity.tsv",
                         sep="\t", na_values="N/A").sort("contig").set_index("contig")
rns31purdf.head()
````

````python
feat_pur = cfeat_counts.join(rns31purdf.ix[:,["contig_length","max_aln_purity"]]).join(rns31cov.ix[:,["GC","cov_mean_sample_0","percentage_covered_sample_0"]])
````

````python
feat_pur.head()
````

````python
f, ax = plt.subplots(1, 1)
badfeat = (feat_pur["feature"]) > 10
longc = feat_pur["contig_length"] >= 1000
ax.hist([feat_pur[longc & badfeat]["max_aln_purity"],
         feat_pur[longc & ~badfeat]["max_aln_purity"]],
         weights=[feat_pur[longc & badfeat]["contig_length"],
                  feat_pur[longc & ~badfeat]["contig_length"]], log=1, range=[0.0,1.0], label=["many features","few features"],
         align='mid')
ax.legend()
ax.set_title("raynoscaf31")
ax.set_ylabel("log # bases in contig with purity (x)")
ax.set_xlabel("purity")
ax.set_ylim(1,1e09)
ax.set_xlim(0.0, 1.0)
````

Using FRC it is still hard to separate pure contigs form impure ones. Maybe there are certain features that we should give more weight or perhaps we should look at the number of bases the features actually cover.

````python
cfeat["feat_nr_bases"] = (cfeat["end"] - cfeat["start"])
````

````python
cfeat_sumbases = cfeat.groupby('contig')["feat_nr_bases"].sum()
cfeat_sumbases.head()
````

````python
feat_pur2 = feat_pur.join(cfeat_sumbases)
feat_pur2.head()
````

````python
f, ax = plt.subplots(1, 1)
badfeat = (feat_pur2["feat_nr_bases"] * 100 / feat_pur["contig_length"]) > 10
longc = feat_pur2["contig_length"] >= 1000
ax.hist([feat_pur2[longc & badfeat]["max_aln_purity"],
         feat_pur2[longc & ~badfeat]["max_aln_purity"]],
         weights=[feat_pur2[longc & badfeat]["contig_length"],
                  feat_pur2[longc & ~badfeat]["contig_length"]], log=1, range=[0.0,1.0], label=["many features","few features"],
         align='mid')
ax.legend()
ax.set_title("raynoscaf31")
ax.set_ylabel("log # bases in contig with purity (x)")
ax.set_xlabel("purity")
ax.set_ylim(1,1e09)
ax.set_xlim(0.0, 1.0)
````

Still no better results than selecting on coverage mean and percentage covered.

````python
mpl.rcParams['figure.figsize'] = 10, 8

mg_length = np.sum(m_even.avp.velvetnoscaf31.genome_contig_coverage["genome_length"])
f, axes = plt.subplots(2, 2, sharex=True, sharey=True)
#ax1.set_xlim([0, 600000])
#ax1.set_title("filtered-refs-filtered-pairs-noscaf")
def plot_metagenome_cov(ax, masm, strategies, category, colors):
    for i, s in enumerate(strategies):
        ax.plot(masm.avp.__dict__[s].contig_metagenome_coverage_purest_single_aln["cut_off"],
                masm.avp.__dict__[s].contig_metagenome_coverage_purest_single_aln["metagenome_cov_percentage"],
                colors[i], linewidth=2.0, label=s)
        ax.plot(masm.avp.__dict__[s].contig_metagenome_coverage_pure_only_single_aln["cut_off"],
                masm.avp.__dict__[s].contig_metagenome_coverage_pure_only_single_aln["metagenome_cov_percentage"],
                colors[i] + "--", linewidth=2.0, label=s)
    ax.set_title(category)

plot_metagenome_cov(axes[0][0], m_even, ['velvetnoscaf31', 'raynoscaf31', 'metavelvetnoscaf31'], 
                    'noscaf', ['k','r','g'])
plot_metagenome_cov(axes[0][1], m_even, ['velvetscaf31', 'rayscaf31', 'metavelvetscaf31'], 
                    'scaf', ['k','r','g'])
plot_metagenome_cov(axes[1][0], m_even, ['velvetnoscafminimus2', 'raynoscafminimus2', 'metavelvetnoscafminimus2'], 
                    'minimus2', ['k','r','g'])
plot_metagenome_cov(axes[1][1], m_even, ['velvetnoscafnewbler', 'raynoscafnewbler', 'metavelvetnoscafnewbler'], 
                    'newbler', ['k','r','g'])



for ax in axes.flatten():
    ax.set_ylim([0,100])
    ax.set_xlim([10**6,0])
    
for ax in axes.flatten():
    ax.set_xscale('log')
    ax.invert_xaxis()
    handles, labels = ax.get_legend_handles_labels()
    handles = [h for i, h in enumerate(handles) if i % 2 == 0]
    labels = [l for i, l in enumerate(labels) if i % 2 == 0]
    ax.legend(handles, labels, numpoints=1, loc='upper left')

#plt.ylabel('Metagenome coverage (%)')
#plt.xlabel('Contig/Scaffold length')

# Legend
# Set common labels
f.text(0.5, 0.08, 'Contig/Scaffold length', ha='center', va='center')
f.text(0.08, 0.5, 'Metagenome coverage (%)', ha='center', va='center', rotation='vertical')
````
