````python
def count_values(test_df, column):
    hist_col = {}
    vals = test_df[column].unique()
    vals.sort()
    for val in  vals:
        if val < max_val:
            hist_col[int(val)] = len(test_df[column][test_df[column] == val])
        else:
            #print test_df[column][test_df[column] >= val], val, max_val, column
            hist_col["> {0}".format(max_val)] = len(test_df[column][test_df[column] >= val].index)
            return(hist_col)
    return hist_col
````
