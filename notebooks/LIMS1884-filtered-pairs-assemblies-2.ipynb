````python
sns.set_style("white", {'xtick.major.size': 5, 'ytick.major.size': 5})
mpl.rcParams['figure.figsize'] = 6, 4

f, ax = plt.subplots(1, 1)

cov_intervals = []
cov_weights = []
ivs = [(0, 25), (25, 50), (50, 75), (75,100), (100, 125), (100, 150), (150, float("inf"))]
for l, h in ivs:
    covi = np.array(rns31cov["cov_mean_sample_0"] >= l) & np.array(rns31cov["cov_mean_sample_0"] < h) & np.array(rns31pur["max_aln_purity"] > 0)
    cov_intervals.append(list(rns31pur[covi]["max_aln_purity"]))
    cov_weights.append(list(rns31pur[covi]["contig_length"]))
ax.hist(cov_intervals,
         weights=cov_weights, log=1, range=[0.0,1.0], label=["{}-{}".format(l,h) for l,h in ivs],
         align='mid', color=sns.dark_palette("palegreen", len(ivs)))
ax.legend()
ax.set_title("raynoscaf31")
ax.set_ylabel("log # bases in contig with purity (x)")
ax.set_xlabel("purity")
ax.set_ylim(1,1e09)
ax.set_xlim(0.0, 1.0)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, numpoints=1, loc='upper left')
ax.yaxis.grid(True, c='white')
ax.set_axisbelow(False)
sns.despine()
````

In these plots the light lines are the metagenome coverage using only 100% pure contigs.  As we can see, the merging of assemblies using newbler and minimus2 provides longer contigs while retaining much of the metagenome coverage. It would seem that by discarding unpure contigs you would lose a lot of coverage, but actually most of the contigs are in the 90-100% purity range (see [below](#Varying-purity-thresholds)) so if you are willing to sacrifice a certain amount of purity for longer contigs then it'll be the best option.
