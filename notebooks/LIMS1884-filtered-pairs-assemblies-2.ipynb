#### Extracting pure contigs with FRC
[FRCurve](https://github.com/vezzi/FRC_align) is a referenceless assembly validation tool. It gives you a lot of different ouput files as can be seen in the readme on the github page. We are interested in getting the number of features indicating a wrong assembly per contig. I ran the program for raynoscaf31 with:

## Metagenome coverage

````python
mpl.rcParams['figure.figsize'] = 10, 8

mg_length = np.sum(m_even.avp.velvetnoscaf31.genome_contig_coverage["genome_length"])
f, axes = plt.subplots(2, 2, sharex=True, sharey=True)
#ax1.set_xlim([0, 600000])
#ax1.set_title("filtered-refs-filtered-pairs-noscaf")
def plot_metagenome_cov(ax, masm, strategies, category, colors):
    for i, s in enumerate(strategies):
        ax.plot(masm.avp.__dict__[s].contig_metagenome_coverage_purest_single_aln["cut_off"],
                masm.avp.__dict__[s].contig_metagenome_coverage_purest_single_aln["metagenome_cov_percentage"],
                colors[i], linewidth=2.0, label=s)
        ax.plot(masm.avp.__dict__[s].contig_metagenome_coverage_pure_only_single_aln["cut_off"],
                masm.avp.__dict__[s].contig_metagenome_coverage_pure_only_single_aln["metagenome_cov_percentage"],
                colors[i] + "--", linewidth=2.0, label=s)
    ax.set_title(category)

plot_metagenome_cov(axes[0][0], m_even, ['velvetnoscaf31', 'raynoscaf31', 'metavelvetnoscaf31'], 
                    'noscaf', ['k','r','g'])
plot_metagenome_cov(axes[0][1], m_even, ['velvetscaf31', 'rayscaf31', 'metavelvetscaf31'], 
                    'scaf', ['k','r','g'])
plot_metagenome_cov(axes[1][0], m_even, ['velvetnoscafminimus2', 'raynoscafminimus2', 'metavelvetnoscafminimus2'], 
                    'minimus2', ['k','r','g'])
plot_metagenome_cov(axes[1][1], m_even, ['velvetnoscafnewbler', 'raynoscafnewbler', 'metavelvetnoscafnewbler'], 
                    'newbler', ['k','r','g'])



for ax in axes.flatten():
    ax.set_ylim([0,100])
    ax.set_xlim([10**6,0])
    
for ax in axes.flatten():
    ax.set_xscale('log')
    ax.invert_xaxis()
    handles, labels = ax.get_legend_handles_labels()
    handles = [h for i, h in enumerate(handles) if i % 2 == 0]
    labels = [l for i, l in enumerate(labels) if i % 2 == 0]
    ax.legend(handles, labels, numpoints=1, loc='upper left')

#plt.ylabel('Metagenome coverage (%)')
#plt.xlabel('Contig/Scaffold length')

# Legend
# Set common labels
f.text(0.5, 0.08, 'Contig/Scaffold length', ha='center', va='center')
f.text(0.08, 0.5, 'Metagenome coverage (%)', ha='center', va='center', rotation='vertical')
````
